```python
api="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"
```

```python
prompt = f'"Summarize this "{joined}'
```

```python
from groq import Groq
client = Groq(api_key=api)

completion = client.chat.completions.create(
    model="llama3-8b-8192",
    messages=[
        {
            "role": "user",
            "content": prompt,
        },
    ],
    temperature=1,
    max_tokens=1024,
    top_p=1,
    stream=True,
    stop=None,
)

for chunk in completion:
    print(chunk.choices[0].delta.content or "", end="")
```
